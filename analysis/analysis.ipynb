```python
import matplotlib.pyplot as plt
import numpy as np
from astro import LinearASTRO
```

```python
def plot_linear_astro(linear_astro, time_since_start, max_sold): 
    
    purchase_quantities = range(1, max_sold)
    prices = [linear_astro.get_price(time_since_start, q) for q in purchase_quantities]

    plt.bar(purchase_quantities, prices)
    plt.xlabel("Quantity of Tokens Sold")
    plt.ylabel("Price (ETH)")
    plt.title(f'Linear ASTRO: Purchase Prices (Î»: {linear_astro.period_price_decrease}, T: {time_since_start}, k:{linear_astro.initial_price})')
    plt.xticks(purchase_quantities)
    plt.show()
```

```python
linear_astro = LinearASTRO(initial_price=100, period_price_decrease=0.2, per_day=2)
plot_linear_astro(linear_astro, time_since_start=2, max_sold=10)
```
